```python
cosmetics_df = pd.read_csv("https://raw.githubusercontent.com/evangelialampropoulou/fashion-blog/main/Chemicals_in_Cosmetics.csv")
cosmetics_df
```

```python
cosmetics_df1 = cosmetics_df.drop(columns=['CDPHId', 'CSF' , 'CompanyId' , 'CasId' , 'CSFId' , 'PrimaryCategoryId' , 'SubCategoryId' , '...' , 'CasNumber' , 'ChemicalId' , 'MostRecentDateReported' , 'InitialDateReported'], errors='ignore')
```

```python
cosmetics_df1
```

```python
cosmetics_df1 = cosmetics_df1.dropna()
cosmetics_df1
```

```python
particular_columns = ['PrimaryCategory', 'DiscontinuedDate']
```

```python
primarycategory_df = cosmetics_df1[particular_columns]
```

```python
primarycategory_df
```

```python
primarycategory_df['DiscontinuedDate'] = pd.to_datetime(primarycategory_df['DiscontinuedDate'])
```

```python
primarycategory_df['Discontinue Year'] = primarycategory_df['DiscontinuedDate'].dt.year
```

```python
groupbyprimarycategory_df = primarycategory_df.groupby('PrimaryCategory')['Discontinue Year'].agg(lambda x: list(set(x))).reset_index()
```

```python
groupbyprimarycategory_df = primarycategory_df.groupby('PrimaryCategory')['Discontinue Year'].agg(list).reset_index()
```

```python
groupbyprimarycategory_df['Discontinue Year'] = groupbyprimarycategory_df['Discontinue Year'].apply(lambda x: ', '.join(map(str, x)) if x else None)
```

```python
groupbyprimarycategory_df = primarycategory_df.groupby('PrimaryCategory')['Discontinue Year'].agg(lambda x: ' '.join(map(str, set(x)))).reset_index()
```

```python
groupbyprimarycategory_df
```

```python
groupbyprimarycategory_df.to_csv('groupbyprimarycategory_df.csv', index=False)

#files.download('groupbyprimarycategory_df.csv')
```

```python
grouped_bychemicalname = cosmetics_df1.groupby('ChemicalName').size().reset_index(name='chemical_count')
```

```python
grouped_bychemicalname
```

```python
grouped_bychemicalname = grouped_bychemicalname.sort_values(by='chemical_count', ascending=False)
```

```python
grouped_bychemicalname
```

```python
grouped_bycompanyname = cosmetics_df1.groupby('CompanyName').size().reset_index(name='company_count')
grouped_bycompanyname = grouped_bycompanyname.sort_values(by='company_count', ascending=False)
```

```python
grouped_bycompanyname
```

```python
grouped_byproductname = cosmetics_df1.groupby('ProductName').size().reset_index(name='product_count')
grouped_byproductname = grouped_byproductname.sort_values(by='product_count', ascending=False)
```

```python
combined_df = pd.concat([grouped_bycompanyname, grouped_bychemicalname , grouped_byproductname], axis=1)
```

```python
combined_df
```

```python
combined_df.to_csv('combined_df.csv', index=False)

#files.download('combined_df.csv')
```

```python
adidas_df = pd.read_csv("https://raw.githubusercontent.com/evangelialampropoulou/fashion-blog/main/adidas_usa.csv")
adidas_df
```

```python
grouped_bycountry = adidas_df.groupby('country').size().reset_index(name='country_count')
grouped_bycountry
```

```python
adidas_df1 = adidas_df.drop(columns=['url', 'name' , 'sku' , 'currency' , 'availability' , 'source' , 'source_website' , 'description' , 'brand' , 'images' , 'country' , 'language' , 'crawled_at'], errors='ignore')
```

```python
adidas_df1
```

```python
adidas_df1 = adidas_df1.dropna()
```

```python
adidas_df1
```

```python
adidas_df1['original_price'] = pd.to_numeric(adidas_df1['original_price'], errors='coerce')
```

```python
result_df = adidas_df1.groupby('breadcrumbs').agg(
    category=('category', lambda x: x.mode().iloc[0] if not x.mode().empty else None),
    average_selling_price=('selling_price', 'mean'),
    color=('color', lambda x: x.mode().iloc[0] if not x.mode().empty else None),
    average_rating=('average_rating', 'mean'),
    reviews_count=('reviews_count', 'mean'),
    original_price=('original_price', 'mean')
).reset_index()
```

```python
result_df
```

```python
result_df.to_csv('result_df.csv', index=False)

#files.download('result_df.csv')
```

```python
c2cfashionstore_df = pd.read_csv("https://raw.githubusercontent.com/evangelialampropoulou/fashion-blog/main/6M-0K-99K.users.dataset.public.csv")
c2cfashionstore_df
```

```python
c2cfashionstore_df = c2cfashionstore_df.drop(columns=['identifierHash', 'type' , 'productsPassRate' , 'civilityTitle' , 'hasProfilePicture' , 'daysSinceLastLogin' , 'seniority' , '...' , 'seniorityAsMonths' , 'countryCode' , 'civilityGenderId'], errors='ignore')
```

```python
c2cfashionstore_df
```

```python
grouped_by_country = c2cfashionstore_df.groupby('country').agg(
    language=('language', lambda x: x.mode().iloc[0]),
    socialNbFollowers=('socialNbFollowers', 'mean'),
    socialNbFollows=('socialNbFollows', 'mean'),
    socialProductsLiked=('socialProductsLiked', 'mean'),
    productsListed=('productsListed', 'mean'),
    productsSold=('productsSold', 'mean'),
    productsWished=('productsWished', 'mean'),
    productsBought=('productsBought', 'mean'),
    gender=('gender', lambda x: x.mode().iloc[0]),
    hasAnyApp=('hasAnyApp', lambda x: x.mode().iloc[0]),
    hasAndroidApp=('hasAndroidApp', lambda x: x.mode().iloc[0]),
    hasIosApp=('hasIosApp', lambda x: x.mode().iloc[0]),
    seniorityAsYears=('seniorityAsYears', 'mean')
).reset_index()
```

```python
grouped_by_country
```

```python
grouped_by_country.to_csv('grouped_by_country.csv', index=False)

#files.download('grouped_by_country.csv')
```

```python
plastic_waste_df = plastic_waste_df.drop(columns=['Flag Codes', 'Flags'], errors='ignore')
```

```python
plastic_waste_df
```

```python
groupedplastic_waste_df = plastic_waste_df.groupby('PLASTICS_APPLICATIONS').size().reset_index(name='plastics_application')
```

```python
groupedplastic_waste_df = groupedplastic_waste_df.sort_values(by='plastics_application', ascending=False)
```

```python
groupedplastic_waste_df
```

```python
groupedplastic_waste_df1 = groupedplastic_waste_df[groupedplastic_waste_df['PLASTICS_APPLICATIONS'] != 'TOTAL']
```

```python
groupedplastic_waste_df1 = groupedplastic_waste_df[groupedplastic_waste_df['PLASTICS_APPLICATIONS'] != 'OTHER']
```

```python
groupedplastic_waste_df1 = groupedplastic_waste_df1.sort_values(by='plastics_application', ascending=False)
```

```python
groupedplastic_waste_df1
```

```python
groupedplastic_waste_df1['PLASTICS_APPLICATIONS'] = groupedplastic_waste_df1['PLASTICS_APPLICATIONS'].replace(['TEXTILE_SECTOR_-_CLOTHING', 'TEXTILE_SECTOR_-_OTHERS'], 'TEXTILE_SECTOR_-_CLOTHING_N_OTHERS')
```

```python
groupedplastic_waste_df1 = groupedplastic_waste_df1.groupby('PLASTICS_APPLICATIONS')['plastics_application'].sum().reset_index()
```

```python
groupedplastic_waste_df1 = groupedplastic_waste_df1.sort_values(by='plastics_application', ascending=False)
```

```python
groupedplastic_waste_df1
```

```python
groupedplastic_waste_df1 = groupedplastic_waste_df1[~groupedplastic_waste_df1['PLASTICS_APPLICATIONS'].isin(['OTHER', 'TOTAL'])]
```
